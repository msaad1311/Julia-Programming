```julia
using ScikitLearn
@sk_import linear_model: LogisticRegression
using DataFrames
using RDatasets: dataset
using Plots
using StatsPlots
```

```julia
data = dataset("datasets","iris")
print(first(data,5))
```

```julia
@df data scatter(
    :SepalLength,
    :SepalWidth,
    group = :Species,
    xlabel = "Length",
    ylabel = "Width",
    title = "Exploratory Analysis"
)
```

```julia
x = convert(Array,data[:,[:SepalLength,:SepalWidth,:PetalLength,:PetalWidth]])
y = convert(Array,data[!,"Species"])
print("Variables Loaded")
```

```julia
print("The dimension of input variable is $(size(x)) and the dimensions of output variable is $(size(y))")
```

```julia
model = LogisticRegression(fit_intercept=true,random_state=42,max_iter=200)
fit!(model,x,y)
```

```julia
accuracy = sum(predict(model, x) .== y) / length(y)
println("accuracy: $accuracy")
```
